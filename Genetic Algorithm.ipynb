```python
# step 2-1: completion time for the Fitness Evaluation
# the function returns the completion time for each chromosome
def completion_time(mydf, chromosome):

  mydf = reorder(mydf, chromosome)

  time = 0
  time = sum(mydf['Prime_T'][i] for i in range(2))

  for i in range(2, 20):
      time += max(mydf['Prime_T'][i-2], mydf['Prime_T'][i-1], mydf['Prime_T'][i])

  # RackID 0 in rack 46
  time += max(mydf['Prime_T'][18], mydf['Prime_T'][19], mydf['Prime_T'][20], mydf['TopCoat_T'][0])
  # RackID 0 in rack 47
  time += max(mydf['Prime_T'][19], mydf['Prime_T'][20], mydf['Prime_T'][21], mydf['TopCoat_T'][0], mydf['TopCoat_T'][1])

  for i in range(142): # i == 142 means the last part in Rack 1
      time += max(mydf['TopCoat_T'][i], mydf['TopCoat_T'][i+1], mydf['TopCoat_T'][i+2], mydf['Prime_T'][i+20], mydf['Prime_T'][i+21], mydf['Prime_T'][i+22])

  # the last RackID 163 in rack 27
  time += max(mydf['TopCoat_T'][142], mydf['TopCoat_T'][143], mydf['TopCoat_T'][144], mydf['Prime_T'][162], mydf['Prime_T'][163])

  # the last part RackID 163 in rack 28
  time += max(mydf['TopCoat_T'][143], mydf['TopCoat_T'][144], mydf['TopCoat_T'][145], mydf['Prime_T'][163])

  for i in range(144, 162): # i == 144 means the last part in Rack 29 and the 144th part in Rack 48, i == 160 means the last part in Rack 45
      time += max(mydf['TopCoat_T'][i], mydf['TopCoat_T'][i+1], mydf['TopCoat_T'][i+2])

  # the last part RackID 163 in rack 47
  time += max(mydf['TopCoat_T'][162], mydf['TopCoat_T'][163])

  # the last part RackID 163 in rack 48
  time += mydf['TopCoat_T'][163]

  return time


population_size = 4
population = Generate_Initial_Population(n_racks, population_size)

time = [completion_time(df2, population[i][0]) for i in range(population_size)]
print(time)
#population[0][1] = time[0]
for i in range(population_size):
  population[i][1] = time[i]
  #print(population[i])
```

```python
# step 3: Selection/Parent Function
# to find some solutions from the entire population that we can use to breed new solutions.
# here I used Tournament selection method, i.e., randomly select k chromosomes from the population of size
# return data with the minimum fitness score from the population
def Selection_Function(population) -> list:

    # return random list size population_size/5 from population
    random_k_list = random.sample(population, int(len(population)/2))

    # sort random list using their fitness score
    random_k_list.sort(key = lambda x: x[1])

    # return first element after sort
    return random_k_list[0]

print(Selection_Function(pop1))
```

```python
# step 4: Crossover function
# generate offspring chromosomes
# basic idea is to get 2 chromosomes from the Selection Function and find a way to “mate” them together to form new chromosomes.
# The way here is to exchange values at a particular index based on a random coin toss.
# So if your coin is at 1, you do nothing, if it’s at 0, you swap the values between the parent chromosomes at that particular index.
!pip install iteration_utilities
from iteration_utilities import duplicates

def Crossover_Function(data1, data2):

    # for this function, I modified the uniform crossover function to take care of duplicates after crossover.

    data1[1] = 0
    data2[1] = 0
    chromosome1 = data1[0].copy()
    chromosome2 = data2[0].copy()


    # for each index in both chromosomes, use a coin toss to determine which index is crossed over
    for i in range(len(chromosome1)):

        cointoss = random.randrange(2)
        if cointoss == 0:
            chromosome1[i], chromosome2[i] = chromosome2[i], chromosome1[i]

    # find duplicates after crossing over
    dupes_in_ch1 = list(duplicates(chromosome1))
    dupes_in_ch2 = list(duplicates(chromosome2))


    # handle duplicates if any are found
    for i in dupes_in_ch1:
        if i in chromosome1: chromosome1.remove(i)
        chromosome2.append(i)

    for i in dupes_in_ch2:
        if i in chromosome2: chromosome2.remove(i)
        chromosome1.append(i)

    # replaced the modified chromosomes in the data
    data1[0] = chromosome1
    data2[0] = chromosome2

    data1[1] = completion_time(df2, chromosome1)
    data2[1] = completion_time(df2, chromosome2)

    return [data1, data2]

```

```python
# step 5: Mutation Function
# use the swap mutation method that swaps the values at 2 random indexes in the chromosome.
# it returns the Modified data
def Mutation_Function(data) -> list:

    chromosome = data[0]

    randomNum1 = random.randint(0, len(chromosome) - 1)
    randomNum2 = random.randint(0, len(chromosome) - 1)

    # exchange values at 2 random indexes
    chromosome[randomNum1], chromosome[randomNum2] = chromosome[randomNum2], chromosome[randomNum1]

    return data
```
